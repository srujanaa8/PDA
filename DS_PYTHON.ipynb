```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SATSCORES')
plt.ylabel('FREQUENCY')
plt.title('HISTOGRAM OF SAT')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.show()
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT Scores')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.xlabel('SAT Scores')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Scores vs. Acceptance Rate')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'rs')
plt.xlabel('SAT Scores')
plt.ylabel('Top 10% High School Rank')
plt.title('Scatter Plot of SAT Scores vs. Top 10% High School Rank')
plt.show()
```
